<h3>Divorce Rates EDA<h3>

Kazakhstan reported the second highest divorce rate in 2021 after the Maldives. This EDA will look at the divorce data from the World Population Review and see if there are determinant of divorce. Additional data was collected from UN sites and Kaggle. I used my previous EDAs on happyness score and gender inquality index, which included health, education, labor, economics and political indicators. I added a dataset called Countries Different Parameters which included yearly population growth, migrant population, fertility age and household size. I used Radom Forest algorithm for feature importance that predict divorce rates in countries. For visualization I used percent stacked bar charts to demonstrate devorce rate and marrige by geographic regions.

<h4>Random Forest<h4>

<h5> Standartization<h5>
    

```python
rfr = RandomForestRegressor(n_estimators=20, max_features=1, max_depth=3, n_jobs=-1, random_state=42)
```

```python
X = div_all.select_dtypes('number').fillna(-999).drop(columns=['divorceRate'])
y = div_all.divorceRate
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

```python
rfr.fit(X_train, y_train)
print('train:', rfr.score(X_train, y_train))
print('test:', rfr.score(X_test, y_test))
```

```python
inspector=get_inspector(rfr.estimators_[0], X, y)
ax=inspector.show_model(max_depth=3)
```

```python
rfr.feature_importances_
```

```python
#pd.Series(grid.best_estimator_.feature_importances_, X.columns).sort_values().plot.barh()
div_feat=pd.DataFrame(rfr.feature_importances_, X_train.columns)
div_feat.columns=['importance']
div_feat=div_feat.sort_values(by='importance', ascending=False).reset_index(drop=False)
div_feat['cumulative_importance']=div_feat['importance'].cumsum()
div_feat
```

```python
pd.Series(rfr.feature_importances_, X_train.columns).sort_values().plot.barh()
```

```python
grid = GridSearchCV(
    estimator=RandomForestRegressor(n_jobs=-1, random_state=1),
    param_grid={'n_estimators': [7, 10, 20, 100],
               'max_features': [1.0, 0.5, 0.3],
               'max_depth':[3, 5, 10]},
    cv=KFold(5, shuffle=True, random_state=1),
    scoring='neg_mean_absolute_error',
    return_train_score=True,
    verbose=True
)
grid.fit(X_train, y_train)
```

```python
grid.best_score_, grid.best_params_
```

```python
div_all=pd.merge(div_all, hap[['country', 'Regional indicator']], on='country')
```

```python
group_data=div_all.groupby('Regional indicator').agg({
    'divorceRate':'mean', 
    'GII':'mean',
    'happiness_score':'mean',
    'growthRate':'mean',
    'density':'mean',
    'household_size':'mean',
    'Social support':'mean',
    'Freedom to make life choices':'mean',
    'Generosity':'mean',
    'Perceptions of corruption':'mean',
    "Fert Rate":'mean',
    'Med Age':'mean',
    'Urban':'mean'
}).reset_index()
```

```python
group_data
```

```python
keyword_mappings = {
    'Europe': 'Europe',
    'Asia': 'Asia',
    'States':'CIS',
    'Latin America and Caribbean':'LAC',
    'Middle East and North Africa': 'MENA',
    'North America and ANZ':'NA & ANZ',
    'Africa':'Africa'
}

def assign_region(row):
    for keyword, region in keyword_mappings.items():
        if keyword in row['Regional indicator']:
            return region
    return None  

group_data['Region'] = group_data.apply(assign_region, axis=1)
```

```python
group_data
```

```python
group_data_new=group_data.groupby('Region').agg({
    'divorceRate':'mean', 
    'GII':'mean', 
    'happiness_score':'mean', 
    'household_size':'mean',
}).reset_index().sort_values('divorceRate', ascending=False)
```

```python
group_data_new
```

```python
barWidth = 0.25
 
bars1 = group_data_new['divorceRate']
bars2 = group_data_new['GII']
bars3 = group_data_new['happiness_score']
bars4 = group_data_new['household_size']
 
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
 
plt.bar(r1, bars1, color='#a70000', width=barWidth, edgecolor='white', label='var1')
plt.bar(r2, bars2, color='#339900', width=barWidth, edgecolor='white', label='var2')
plt.bar(r3, bars3, color='#005073', width=barWidth, edgecolor='white', label='var3')
plt.bar(r4, bars4, color='#9c7248', width=barWidth, edgecolor='white', label='var3')
 
plt.xlabel('Regions', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['CIS', 'NA&ANZ', 'Asia', 'Europe', 'MENA', 'Africa', 'LAZ'])
 
plt.legend()
plt.show()
```
