```python
div_rates=pd.read_excel('../final_proposal_data/Divorce Rates.xlsx')
```

```python
div_rates_reg=div_suic_gii_hap_para_hou.merge(div_rates, on='country', how='left')[[
    'country', 'Regional indicator', 'Marriage', 'divorceRate']]
```

```python
div_rates_reg
```

```python
div_rates_group=div_rates_reg.groupby('Regional indicator').agg({
    'Marriage':'mean', 'divorceRate':'mean'}).reset_index().sort_values('divorceRate', ascending=False)
```

```python
div_rates_group
```

```python
div_rates_group.columns
```

```python
region_mapping = {
    'Commonwealth of Independent States': 'CIS',
    'North America and ANZ': 'NA & ANZ',
    'Western Europe': 'W. Europe',
    'East Asia': 'E. Asia',
    'Middle East and North Africa': 'MENA',
    'Central and Eastern Europe': 'CEE',
    'Southeast Asia': 'SE Asia',
    'Sub-Saharan Africa': 'Sub-Saharan Africa',
    'Latin America and Caribbean': 'LAC'
}
div_rates_group['Regional indicator'] = div_rates_group['Regional indicator'].map(region_mapping)
```

```python
div_rates_group
```

```python
total = div_rates_group['Marriage']
div_rates_group['Marriage %'] = ((div_rates_group['Marriage']-div_rates_group['divorceRate'])/div_rates_group['Marriage']) * 100
div_rates_group['Divorce Rate %'] = (div_rates_group['divorceRate'] / div_rates_group['Marriage']) * 100

fig, ax = plt.subplots()

ax.bar(div_rates_group['Regional indicator'], div_rates_group['Marriage %'], label="Marriage Rate", alpha=0.8, height=10)

ax.bar(div_rates_group['Regional indicator'], div_rates_group['Divorce Rate %'], bottom=div_rates_group['Marriage %'], label="Divorce Rate", alpha=0.8)

ax.set_xlabel("Regional Indicator")
ax.set_ylabel("Percentage")
ax.set_title("Marriage and Divorce Rates by Regional Indicator")
#ax.set_xticks(index)
ax.set_xticklabels(div_rates_group['Regional indicator'], rotation=90)
ax.legend()

plt.tight_layout()
plt.show()
```
